```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\Users\abdal\Desktop\archive_3 (1)/books.csv')
```

```python
# Replace missing values with empty string in 'original_title' column
df['original_title'].fillna('', inplace=True)
```

```python
# Filter books with "Harry Potter" in the title
# Filter rows for Harry Potter books
harry_potter_books = df[df['original_title'].str.contains('Harry Potter')]
```

```python
# Calculate the average rating
average_rating = harry_potter_books['average_rating'].mean()
```

```python
print("Average rating of Harry Potter books:", average_rating)
```

```python
# Sort the books by the number of ratings_count in descending order
sorted_books = harry_potter_books.sort_values(by='ratings_count', ascending=False)
```

```python
# Get the top selling book
most_selling_book = sorted_books.iloc[0]
```

```python
# Print the details of the most selling book
print("Most Selling Book in the Harry Potter Series:")
print("Title:", most_selling_book['original_title'])
print("Author:", most_selling_book['authors'])
print("Total Ratings Count:", most_selling_book['ratings_count'])
```

```python
df = df.drop(['best_book_id', 'work_id', 'isbn', 'isbn13', 'image_url', 'small_image_url'], axis=1)
```

```python
df.isnull().sum()  # Check for missing values in each column
df = df.dropna()  # Drop rows with missing values
```

```python
df['original_publication_year'] = df['original_publication_year'].astype(int)
```

```python
harry_potter_books = df[df['original_title'].str.contains('Harry Potter')]
```

```python
sns.histplot(data=harry_potter_books, x='average_rating', bins=10)
plt.title('Distribution of Ratings for Harry Potter Books')
plt.show()
```

```python
sns.barplot(data=harry_potter_books, x='original_title', y='ratings_count')
plt.title('Ratings Count for Harry Potter Books')
plt.xticks(rotation=45)
plt.show()
```

```python
hp_publication_trend = harry_potter_books.groupby('original_publication_year').size()
hp_publication_trend.plot(kind='line', marker='o')
plt.title('Publication Trend of Harry Potter Books')
plt.xlabel('Year')
plt.ylabel('Number of Books')
plt.show()
```
